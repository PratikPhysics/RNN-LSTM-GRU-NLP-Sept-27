```python
# ============================
# NIFTY Price Prediction Teaching Notebook
# RNN / LSTM / GRU — many variants, visuals & comparisons
# Uses yfinance to download data
# ============================

import yfinance as yf
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
import time, math
from sklearn.preprocessing import MinMaxScaler
from sklearn.metrics import mean_squared_error, mean_absolute_error
from statsmodels.tsa.seasonal import seasonal_decompose   # optional, good for teaching
import tensorflow as tf
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import Dense, Dropout, SimpleRNN, LSTM, GRU
from tensorflow.keras.callbacks import EarlyStopping

# -----------------------
# Reproducibility (good for teaching)
# -----------------------
SEED = 42
np.random.seed(SEED)
tf.random.set_seed(SEED)

# -----------------------
# Configuration - change these to demonstrate effects
# -----------------------
TICKER = "^NSEI"          # NIFTY 50 on Yahoo Finance
START_DATE = "2010-01-01"
END_DATE = None           # None => up to today
USE_FEATURES = ["Close"]  # ["Close"] or ["Open","High","Low","Close","Volume"]
TIME_STEPS = 60           # sequence length (window)
TEST_SPLIT = 0.2
VIZ_SAMPLE_DAYS = 1000    # how many recent days to show in some plots
EPOCHS = 20               # smaller for demo, increase for more training
BATCH_SIZE = 32

# -----------------------
# 1) Download Data with yfinance
# -----------------------
print("Downloading data for", TICKER)
df = yf.download(TICKER, start=START_DATE, end=END_DATE, progress=False)
if df.empty:
    raise RuntimeError("No data downloaded. Check ticker or date range.")
df = df.reset_index()
df['Date'] = pd.to_datetime(df['Date'])
df = df.sort_values('Date').reset_index(drop=True)
print("Data downloaded:", df.shape)
print(df.head())

# -----------------------
# Teaching concept:
#   Show raw time series and recent behavior
# -----------------------
plt.figure(figsize=(14,5))
plt.plot(df['Date'], df['Close'], lw=1)
plt.title("NIFTY — Closing Price (full range)")
plt.xlabel("Date"); plt.ylabel("Close Price")
plt.show()

plt.figure(figsize=(14,5))
plt.plot(df['Date'].iloc[-VIZ_SAMPLE_DAYS:], df['Close'].iloc[-VIZ_SAMPLE_DAYS:], lw=1)
plt.title(f"NIFTY — Closing Price (last {VIZ_SAMPLE_DAYS} days)")
plt.xlabel("Date"); plt.ylabel("Close Price")
plt.show()

# -----------------------
# 2) Teach: Decomposition (trend/seasonality/residual)
# -----------------------
try:
    decomp_period = 252  # ~ trading days in a year
    decomposed = seasonal_decompose(df['Close'], period=decomp_period, model='multiplicative', extrapolate_trend='freq')
    fig = decomposed.plot()
    fig.set_size_inches(14,10)
    plt.suptitle("Seasonal Decompose (multiplicative) — Teaching visual")
    plt.show()
except Exception as e:
    print("Seasonal decomposition skipped (reason):", e)

# -----------------------
# 3) Preprocessing: choose features, scale
#    Optionally you can set USE_FEATURES to include OHLCV for multivariate models
# -----------------------
features = USE_FEATURES.copy()
data = df[features].astype('float32').values
print("Using features:", features)

scaler = MinMaxScaler(feature_range=(0,1))
data_scaled = scaler.fit_transform(data)    # for multivariate this scales each feature

# Teaching note:
#   Explain why scaling is required (gradient stability, faster convergence)
#   For univariate vs multivariate differences, show shape below.

print("Scaled data shape:", data_scaled.shape)

# -----------------------
# 4) Create sequences for supervised learning
# -----------------------
def create_sequences_multivariate(data_array, time_steps):
    """
    data_array: numpy array shape (n_samples, n_features)
    returns X shape (n_windows, time_steps, n_features), y shape (n_windows, n_features_to_predict)
    For simplicity we'll predict the same feature set's last timestep value (typically Close)
    """
    X, y = [], []
    for i in range(time_steps, len(data_array)):
        X.append(data_array[i-time_steps:i])
        y.append(data_array[i])   # predict next timestep's all features (we'll typically use only Close when evaluating)
    X = np.array(X)
    y = np.array(y)
    return X, y

X, y = create_sequences_multivariate(data_scaled, TIME_STEPS)
print("Sequence shapes X, y:", X.shape, y.shape)

# For teaching, print an example window (decoded back to original scale for clarity)
sample_idx = -100  # show a sample near the end
example_window = scaler.inverse_transform(X[sample_idx].reshape(-1, X.shape[-1])).flatten()[:10]  # partial print
print("\nExample window (first 10 values in original scale from the last sample):")
print(example_window)

# -----------------------
# 5) Train-test split (time-ordered)
# -----------------------
split_idx = int((1 - TEST_SPLIT) * len(X))
X_train, X_test = X[:split_idx], X[split_idx:]
y_train, y_test = y[:split_idx], y[split_idx:]

print("Train shape:", X_train.shape, "Test shape:", X_test.shape)

# Teaching point:
#   - We DO NOT shuffle because time order matters
#   - Explain walk-forward evaluation, but here we do a single train/test split for clarity

# -----------------------
# 6) Model builders (progressive units: 128 -> 64 -> 32)
#    Builders return compiled models accepting input_shape = (TIME_STEPS, n_features)
# -----------------------
n_features = X.shape[-1]
early_stop = EarlyStopping(monitor='val_loss', patience=3, restore_best_weights=True)

def build_rnn(units_list=[64], dropout=0.2, input_shape=(TIME_STEPS, n_features)):
    model = Sequential()
    for i, units in enumerate(units_list):
        return_seq = (i < len(units_list)-1)
        if i == 0:
            model.add(SimpleRNN(units, return_sequences=return_seq, input_shape=input_shape))
        else:
            model.add(SimpleRNN(units, return_sequences=return_seq))
        model.add(Dropout(dropout))
    model.add(Dense(n_features))   # predict all features; when evaluating we'll use Close column
    model.compile(optimizer='adam', loss='mse', metrics=['mae'])
    return model

def build_lstm(units_list=[64], dropout=0.2, input_shape=(TIME_STEPS, n_features)):
    model = Sequential()
    for i, units in enumerate(units_list):
        return_seq = (i < len(units_list)-1)
        if i == 0:
            model.add(LSTM(units, return_sequences=return_seq, input_shape=input_shape))
        else:
            model.add(LSTM(units, return_sequences=return_seq))
        model.add(Dropout(dropout))
    model.add(Dense(n_features))
    model.compile(optimizer='adam', loss='mse', metrics=['mae'])
    return model

def build_gru(units_list=[64], dropout=0.2, input_shape=(TIME_STEPS, n_features)):
    model = Sequential()
    for i, units in enumerate(units_list):
        return_seq = (i < len(units_list)-1)
        if i == 0:
            model.add(GRU(units, return_sequences=return_seq, input_shape=input_shape))
        else:
            model.add(GRU(units, return_sequences=return_seq))
        model.add(Dropout(dropout))
    model.add(Dense(n_features))
    model.compile(optimizer='adam', loss='mse', metrics=['mae'])
    return model

# -----------------------
# 7) Define model variants (progressive reduction)
# -----------------------
configs = [
    ("RNN-32",       build_rnn([32], dropout=0.1)),
    ("RNN-64-32",    build_rnn([64,32], dropout=0.2)),
    ("RNN-128-64-32",build_rnn([128,64,32], dropout=0.3)),
    ("LSTM-32",       build_lstm([32], dropout=0.1)),
    ("LSTM-64-32",    build_lstm([64,32], dropout=0.2)),
    ("LSTM-128-64-32",build_lstm([128,64,32], dropout=0.3)),
    ("GRU-32",       build_gru([32], dropout=0.1)),
    ("GRU-64-32",    build_gru([64,32], dropout=0.2)),
    ("GRU-128-64-32",build_gru([128,64,32], dropout=0.3)),
]

# Teaching tip:
#   - Use smaller models first so students see results quickly
#   - Then demonstrate larger models and show runtime differences

# -----------------------
# 8) Train all variants and record metrics & runtime
# -----------------------
history = {}
results = {}
timings = {}
predictions = {}

for name, model in configs:
    print("\n==============================================")
    print("Training model:", name)
    print(model.summary())
    start = time.time()
    h = model.fit(
        X_train, y_train,
        validation_split=0.15,   # internal val split for early stopping
        epochs=EPOCHS,
        batch_size=BATCH_SIZE,
        callbacks=[early_stop],
        verbose=1
    )
    elapsed = time.time() - start

    # Predict and rescale target (we predict all features but evaluate Close)
    y_pred_scaled = model.predict(X_test)
    # inverse transform: scaler expects shape (n_samples, n_features) -> we already have that
    y_test_rescaled = scaler.inverse_transform(y_test)
    y_pred_rescaled = scaler.inverse_transform(y_pred_scaled)

    # Evaluate on Close column (index of Close in features)
    close_idx = features.index("Close") if "Close" in features else 0
    rmse = math.sqrt(mean_squared_error(y_test_rescaled[:, close_idx], y_pred_rescaled[:, close_idx]))
    mae = mean_absolute_error(y_test_rescaled[:, close_idx], y_pred_rescaled[:, close_idx])

    print(f"{name} RMSE (Close): {rmse:.4f}, MAE (Close): {mae:.4f}, Time: {elapsed:.1f}s")

    history[name] = h
    results[name] = {"rmse": rmse, "mae": mae}
    timings[name] = elapsed
    predictions[name] = {"y_true": y_test_rescaled[:, close_idx], "y_pred": y_pred_rescaled[:, close_idx]}

# -----------------------
# 9) Visual comparisons: loss curves and runtime
# -----------------------
plt.figure(figsize=(14,6))
for name, h in history.items():
    if "val_loss" in h.history:
        plt.plot(h.history['val_loss'], label=name)
plt.title("Validation Loss (MSE) per Model Variant")
plt.xlabel("Epoch")
plt.ylabel("Val Loss")
plt.legend()
plt.show()

plt.figure(figsize=(14,6))
for name, h in history.items():
    if "val_mae" in h.history:
        plt.plot(h.history['val_mae'], label=name)
plt.title("Validation MAE per Model Variant")
plt.xlabel("Epoch")
plt.ylabel("Val MAE")
plt.legend()
plt.show()

# Runtime bar chart
plt.figure(figsize=(10,6))
names = list(timings.keys())
times = [timings[n] for n in names]
sns.barplot(x=times, y=names)
plt.xlabel("Training Time (s)")
plt.title("Training Time per Model Variant")
plt.show()

# -----------------------
# 10) Show predictions vs actual for best model
# -----------------------
best_name = min(results.keys(), key=lambda k: results[k]['rmse'])
print("\nBest model by RMSE (Close):", best_name, results[best_name])

y_true = predictions[best_name]['y_true']
y_pred = predictions[best_name]['y_pred']

plt.figure(figsize=(14,6))
plt.plot(y_true, label='Actual Close (Test)')
plt.plot(y_pred, label='Predicted Close (Test)')
plt.title(f"Actual vs Predicted Close Prices — {best_name}")
plt.xlabel("Test sample index")
plt.ylabel("Price")
plt.legend()
plt.show()

# Zoom into the last 200 test points for clarity
zoom_n = min(200, len(y_true))
plt.figure(figsize=(14,6))
plt.plot(y_true[-zoom_n:], lw=1.5, label='Actual (last points)')
plt.plot(y_pred[-zoom_n:], lw=1.2, label='Predicted (last points)')
plt.title(f"Zoom (last {zoom_n}) — {best_name}")
plt.legend()
plt.show()

# -----------------------
# 11) Error distribution (residuals)
# -----------------------
residuals = y_true - y_pred
plt.figure(figsize=(10,4))
sns.histplot(residuals, bins=50, kde=True)
plt.title(f"Residual Distribution — {best_name}")
plt.xlabel("Residual (Actual - Predicted)")
plt.show()

# -----------------------
# 12) Summary table
# -----------------------
summary_df = pd.DataFrame([
    {"Model": name, "RMSE": results[name]['rmse'], "MAE": results[name]['mae'], "Training Time (s)": round(timings[name],1)}
    for name in results
]).sort_values("RMSE")
print("\n=== Summary Table (sorted by RMSE) ===")
display(summary_df)   # in notebooks; if not available use print(summary_df)

# -----------------------
# Teaching notes & next steps suggestions
# -----------------------
print("""
Teaching progression notes you can use in class:
1) Start with raw plot + decomposition -> motivate stationarity/trend/seasonality.
2) Show the effect of scaling and why sequences are needed.
3) Train small RNN/LSTM/GRU (32 units) — quick results to show model shapes.
4) Increase depth & units (64->32, 128->64->32) and discuss capacity vs overfitting.
5) Show validation loss curves and explain EarlyStopping.
6) Show runtime differences (RNN < GRU < LSTM typically).
7) Demonstrate forecasting quality (RMSE/MAE) and error/residual analysis.
8) Advanced: include exogenous features (Volume, Open/High/Low) for multivariate forecasting,
   or implement walk-forward validation / rolling origin evaluation for more realistic assessment.

Extras you can add in class:
- Grid search over TIME_STEPS, BATCH_SIZE, or units.
- Walk-forward cross-validation.
- Use teacher forcing (sequence-to-sequence) for multi-step forecasts.
- Compare with simple baselines (persistence/ARIMA) to show when deep models add value.
""")
```
